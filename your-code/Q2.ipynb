```python
import re
def remove_punctuation(dirty_doc):
    '''This fin removes everything that is not a letter or number'''
    clean_doc = re.sub(r'[^\w\s]', ' ', dirty_doc)
    return clean_doc
```

```python
def strip_html_tags(dirty_doc):
    '''This func cleans html tags'''
    clean_doc = re.sub(r'<.*?>','',dirty_doc)
    return clean_doc
```

```python
def double_space(dirty_doc):
    '''This func deletes double or more  spacing'''
    clean_doc = re.sub(r' +',' ', dirty_doc)
    return clean_doc
```

```python
def to_lower_case(dirty_doc):
    '''This func converts everything to lower case'''
    clean_doc = re.sub(r'[A-Z]+','\[a-z]', dirty_doc)
    return clean_doc
```

```python
def remove_unicode(dirty_doc):
    '''This func removes all unicode'''
    clean_doc = re.sub(r'[^[A-Za-z0-9]]', '', dirty_doc)
    return clean_doc
```
